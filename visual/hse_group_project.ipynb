```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
```

```python
df = pd.read_csv(r"C:\Users\Alex\Downloads\clusters_with_z_score.tsv", delimiter="\t")
```

```python
df
```

```python
df_new = df.iloc[:,52:-1].replace("*", 0).astype('float')[:20]
df_new.shape
```

```python

cmap =  sns.color_palette("YlOrRd", as_cmap=True)
cmap.set_under('gray')
fig, ax = plt.subplots()
fig.set_size_inches(15, 10)
sns.heatmap(df_new.T, vmax = 10000, vmin=500, linewidth=0.05, cmap=cmap)
plt.xlabel("Clusters")
plt.ylabel("Organisms")
plt.savefig("clusters.jpg")
```
